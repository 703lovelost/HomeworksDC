## Задание 6.

Помимо установки пакетов на Linux вы также можете устанавливать пакеты для языка программирования Python - отдельные библиотеки, открывающие новый функционал для разработки.

Для следующего эксперимента установим пакет [matplotlib](https://pypi.org/project/matplotlib/), используя [pip](https://pypi.org/project/pip/), менеджер пакетов для Python.

```python
!pip install matplotlib
```

В коде ниже вставьте любую прямую ссылку на картинку в строчку url. Запустите ячейку несколько раз с разными картинками.

```python
import matplotlib.pyplot as plt
from io import BytesIO
from PIL import Image
import requests

url = "https://external-content.duckduckgo.com/iu/?u=https%3A%2F%2Fwallpapercave.com%2Fwp%2Fwp2754862.jpg&f=1&nofb=1&ipt=aa4e4f1b9a6dd6e0b1ae7183d7b3a68c84ff502a3ce1198f7b7c788d7241b3f3"

resp = requests.get(url, timeout=15)
resp.raise_for_status()
img = Image.open(BytesIO(resp.content)).convert("RGB")

plt.figure(figsize=(6, 6))
plt.imshow(img)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
